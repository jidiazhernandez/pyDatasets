```python
from __future__ import division

import argparse
import glob
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import sys

from datetime import datetime, timedelta
from pandas import Panel, DataFrame

from pydatasets import icu

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 50)
pd.set_option('display.width', 160)
```

```python
parser = argparse.ArgumentParser()
parser.add_argument('data', type=unicode)
parser.add_argument('out', type=unicode)
parser.add_argument('-vi', '--var_info', type=unicode, default='../data-info/picu/physiologic-variable-normals.csv')
parser.add_argument('-ai', '--age_dep_var_info', type=unicode, default='../data-info/picu/physiologic-variable-ageranges.csv')
parser.add_argument('-ar', '--age_ranges', type=unicode, default='../data-info/picu/physiologic-variable-ageranges.csv')
parser.add_argument('-rr', '--resample_rate', type=int, default=60)

import __main__
if not hasattr(__main__, '__file__'):
    import platform
    userdir = 'Users' if platform.system()=='Darwin' else 'home'
    args = parser.parse_args(args=[ '/' + userdir + '/davekale/mldata/icu/dataset1.0-phi-pickled-201410311028', 
                                    '/' + userdir + '/davekale/mldata/icu', '-rr', '60' ])
else:
    args = parser.parse_args()

dts = dts = datetime.now().strftime('%Y%m%d%H%M')
    
data_dir = args.data
out_dir = args.out
try:
    os.makedirs(out_dir)
except:
    pass

# Get (Low, High) range and Normal value for each variable. For age-dependent variables, this is the "adult female" normal.
ranges = DataFrame.from_csv(args.var_info, parse_dates=False)
ranges = ranges[['Low', 'Normal', 'High']]
varids = ranges.index.tolist()

# Get age/gender-dependent ranges for age-dependent variables (HR, BP, RR).
# We will use these to "adjust" each patient to be on the same scale.
age_dep_ranges = DataFrame.from_csv(args.age_dep_var_info, index_col=None)
dupedata = age_dep_ranges.ix[age_dep_ranges.Gender.isnull()].copy()
dupedata.Gender = 'F'
age_dep_ranges.Gender[age_dep_ranges.Gender.isnull()] = 'M'
age_dep_ranges.ix[age_dep_ranges.Variable=='HR']
age_dep_ranges = pd.concat((age_dep_ranges, dupedata))
age_dep_ranges.sort(columns=['Variable', 'Gender', 'AgeLow'], inplace=True)
age_dep_ranges.Gender = age_dep_ranges.Gender.apply(lambda s: 1 if s=='F' else 0 if s=='M' else np.nan)
age_dep_ranges = age_dep_ranges.set_index('VariableID')
age_dep_ranges = Panel(dict([ (varid, age_dep_ranges.ix[varid].reset_index()[['Gender','AgeLow','AgeHigh','Low','Normal','High']]) for varid in age_dep_ranges.index.unique() ]))

resample_rate = '{0}min'.format(args.resample_rate)
```

```python
# Read directory contents (should be one entry per patient episode)
fns = glob.glob(os.path.join(args.data, '*.pkl'))
N = len(fns) # this is the number of episodes
epids = np.zeros((N,), dtype=int)

# unnormalized data
XUraw = []  # raw time series (without resampling or imputation)
XUmiss = [] # resampled time series (with imputation)
XU = []     # resampled, imputed time series

# znormalized by age-based ranges
XZraw = []  # raw time series (without resampling or imputation)
XZmiss = [] # resampled time series (without imputation)
XZ = []     # resampled, imputed time series

# scaled by age-based normals
XSraw = []  # raw time series (without resampling or imputation)
XSmiss = [] # resampled time series (without imputation)
XS = []     # resampled, imputed time series

Traw = np.zeros((N,), dtype=int)    # Number of (irregular) samples
T = np.zeros((N,), dtype=int)       # Resampled episode lengths (for convenience)
age = np.zeros((N,), dtype=int)     # Per-episode patient ages in months
gender = np.zeros((N,), dtype=int) # Per-episode patient gender
weight = np.zeros((N,))             # Per-episode patient weight
los = np.zeros((N,))                # Length of stay
lmf = np.zeros((N,))                # Length of stay (Last - First)

ym = np.zeros((N,), dtype=int) # Outcome (mortality: 1 if died, 0 if survived)
yd = np.zeros((N,), dtype=int)  # Primary diagnosis code
pcpc = np.zeros((N,2))          # PCPC codes: admission, change
popc = np.zeros((N,2))          # POPC codes: admission, change
Yd = []                      # Per-episode list of diagnostic codes
```

```python
perc = 0.01
sys.stdout.write('processing episodes\n')
count_old = 0
count_long = 0
count_short = 0
for ix, fn in enumerate(fns):
    try:
        ep = icu.VpicuEpisode.from_pickle(fn)
    except:
        print 'Could not unpickle {0}, skipping...'.format(os.path.split(fn)[1])
        assert(False)
    if ix / len(fns) > perc:
        sys.stdout.write('{0}\n'.format(ix))
        perc += 0.01

    s = ep.as_nparray()
    Traw[ix] = s.shape[0]
    lmf[ix] = (s[:,0].max()-s[:,0].min())/60        
    if ep.age_in_months() > 12*18: # patient is >18 years
        count_old += 1
    elif ep.los_hours() <= 1 or lmf[ix] <= 1:
        count_short += 1

    ### Unadjusted measurements: XUraw/XUmiss/XU are NOT adjusted for age ###
    # store raw time series
    XUraw.append(s)
        
    # resample
    s = ep.as_nparray_resampled(rate=resample_rate)
    XUmiss.append(s)
    
    # resample, impute
    s = ep.as_nparray_resampled(rate=resample_rate, impute=True, normal_values=ranges.Normal)
    if not np.all(~np.isnan(s)):
        print ep.episodeid
        assert(np.all(~np.isnan(s)))
    XU.append(s)
    
    ### Znormalized measurements: XZraw/XZmiss/XZ are znormalized by age ranges ###
    msmts_old = ep.msmts
    ep.msmts = ep.adjust_msmts_for_age_gender_znorm(age_dep_ranges)
    
    # store raw time series
    s = ep.as_nparray()
    XZraw.append(s)
    
    # resample
    s = ep.as_nparray_resampled(rate=resample_rate)
    XZmiss.append(s)
    
    # resample, impute
    s = ep.as_nparray_resampled(rate=resample_rate, impute=True, normal_values=ranges.Normal)
    if not np.all(~np.isnan(s)):
        print ep.episodeid
        assert(np.all(~np.isnan(s)))
    XZ.append(s)
    
    ### Znormalized measurements: XZraw/XZmiss/XZ are znormalized by age ranges ###
    ep.msmts = msmts_old
    ep.msmts = ep.adjust_msmts_for_age_gender_scale(age_dep_ranges)
    
    # store raw time series
    s = ep.as_nparray()
    XSraw.append(s)
    
    # resample
    s = ep.as_nparray_resampled(rate=resample_rate)
    XSmiss.append(s)
    
    # resample, impute
    s = ep.as_nparray_resampled(rate=resample_rate, impute=True, normal_values=ranges.Normal)
    if not np.all(~np.isnan(s)):
        print ep.episodeid
        assert(np.all(~np.isnan(s)))
    XS.append(s)

    ### other stuff: los, length, age, gender, weight, etc. ###
    epids[ix] = ep.episodeid
    los[ix] = ep.los_hours()

    T[ix] = s.shape[0]
    age[ix] = ep.age_in_months()
    gender[ix] = ep.sex
    weight[ix] = ep.weight
    
    ### outcomes: mortality, diagnosis, etc. ###
    ym[ix] = np.nan if ep.died is None else 1 if ep.died else 0
    yd[ix] = ep.prim_diag
    pcpc[ix,:] = np.array([ ep.delta_pcpc(), ep.discharge_pcpc ])
    popc[ix,:] = np.array([ ep.delta_popc(), ep.discharge_popc ])
    Yd.append(np.array(list(ep.diag), dtype=int))

sys.stdout.write('DONE!\n')

print '{0} patients were over 18 years old'.format(count_old)
print '{0} episodes were shorter than 1 hour'.format(count_short)
```

```python
sys.stdout.write('Counting outlier (<min, >max) measurements in original data:\n')
outliers = np.zeros((ranges.shape[0]+1,2), dtype=int)
allmsmts1 = dict()
for vi in ranges.index:
    allmsmts1[vi] = np.hstack([ x[:,vi][~np.isnan(x[:,vi])] for x in XUraw ])
    outliers[vi,0] = np.sum(allmsmts1[vi]<ranges.Low[vi])
    outliers[vi,1] = np.sum(allmsmts1[vi]>ranges.High[vi])
    sys.stdout.write('  var {0:2d}: min:{1} max:{2}\n'.format(vi, outliers[vi,0], outliers[vi,1]))
```

```python
sys.stdout.write('Counting outlier (<min, >max) measurements in znormalized data:\n')
outliers = np.zeros((ranges.shape[0]+1,2), dtype=int)
allmsmts2 = dict()
for vi in ranges.index:
    allmsmts2[vi] = np.hstack([ x[:,vi][~np.isnan(x[:,vi])] for x in XZraw ])
    outliers[vi,0] = np.sum(allmsmts2[vi]<ranges.Low[vi])
    outliers[vi,1] = np.sum(allmsmts2[vi]>ranges.High[vi])
    sys.stdout.write('  var {0:2d}: min:{1} max:{2}\n'.format(vi, outliers[vi,0], outliers[vi,1]))
```

```python
sys.stdout.write('Counting outlier (<min, >max) measurements in scaled data:\n')
outliers = np.zeros((ranges.shape[0]+1,2), dtype=int)
allmsmts3 = dict()
for vi in ranges.index:
    allmsmts3[vi] = np.hstack([ x[:,vi][~np.isnan(x[:,vi])] for x in XSraw ])
    outliers[vi,0] = np.sum(allmsmts3[vi]<ranges.Low[vi])
    outliers[vi,1] = np.sum(allmsmts3[vi]>ranges.High[vi])
    sys.stdout.write('  var {0:2d}: min:{1} max:{2}\n'.format(vi, outliers[vi,0], outliers[vi,1]))
```
