```python
from __future__ import division

import argparse
import numpy as np
import os
import pandas as pd
import re
#import scipy.io as sio
import sys

from datetime import datetime
from pandas import Panel, DataFrame

from pydatasets import icu

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 50)
pd.set_option('display.width', 160)
```

```python
parser = argparse.ArgumentParser()
parser.add_argument('data', type=unicode)
parser.add_argument('out', type=unicode)
parser.add_argument('-vi', '--varids', nargs='+', type=int, default=range(1, 14))

import __main__
if not hasattr(__main__, '__file__'):
    import platform
    userdir = 'Users' if platform.system()=='Darwin' else 'home'
    args = parser.parse_args(args=[ '/' + userdir + '/davekale/mldata/icu/dataset1.0-phi-20130910184316', 
                                    '/' + userdir + '/davekale/mldata/icu', '-rr', '60' ])
else:
    args = parser.parse_args()

dts = dts = datetime.now().strftime('%Y%m%d%H%M')

data_dir = args.data

m = re.search('/.*/([\w\.]+)-[^/]/?', args.data)
if m:
    out_dir = os.path.join(args.out, m.group(1) + '-pickled-' + dts)
else:
    out_dir = os.path.join(args.out, 'icu-pickled-' + dts)
try:
    os.makedirs(out_dir)
except:
    pass
```

```python
# Read directory contents (should be one entry per patient episode)
epdirs = os.listdir(data_dir)
eps = []

ct = 0
ct_nodata = 0
perc = 0.01
sys.stdout.write('reading and pickling episodes\n')
for epdir in epdirs:
    try:
        epid = int(epdir) # VPICU episode is stored in directory named after episode ID
    except:
        print 'Invalid subdirectory {0}, skipping...'.format(epdir)
        continue
    if ct / len(epdirs) > perc:
        sys.stdout.write('{0}\n'.format(ct))
        perc += 0.01
    try:
        ep = icu.OldVpicuEpisode.from_directory(os.path.join(data_dir, epdir), varids=varids)
    except icu.InvalidVpicuDataException as e:
        if e.field == 'msmts' and e.err == 'size': # no measurements data
            ct_nodata += 1
        else:
            sys.stdout.write('\nskipping ' + epdir + ': ' + str(e))
        continue
    
    eps.append(ep)
    ep.to_pickle(path=out_dir)
    ct += 1
    
sys.stdout.write('DONE!\n')

print '{0} episodes had no measurements'.format(ct_nodata)
```
